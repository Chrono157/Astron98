```python
planets_data_new = planets_data.dropna(axis = 0)
planets_data_new.insert(3,'log orbital_period',np.log10(planets_data['orbital_period']))
planets_data_new.insert(5,'log mass',np.log10(planets_data['mass']))
g = sns.FacetGrid(planets_data_new,hue='method')
g.map(sns.scatterplot, 'log orbital_period','log mass')
g.tight_layout()
g.add_legend()
plt.show()
```

```python
#Part 2
plt.figure(figsize=[20,10])
sns.barplot(planets_data_new, x="year", y="number", hue="method",errorbar=None)
plt.show()
```

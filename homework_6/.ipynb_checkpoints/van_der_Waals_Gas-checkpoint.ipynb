```python
import numpy as np
import matplotlib.pyplot as plt
#I referenced numpy documentation mainly for this program
```

```python
P = np.linspace(1,10,100)
V = np.linspace(10,30,100)
P,V = np.meshgrid(P,V)
a = 16.02
b = 0.1124
R = 0.083144
n = 1 #we only have 1 mol of acetone
T = ((P+(a*(n**2/(V**2))))*(V-(n*b)))/(n*R) #I calculated 1 value using desmos and compared it to this. This gives the right answer, at the very least at the origin
```

```python
plt.figure()
plt.imshow(T,extent=[1,10,10,30],cmap="ocean",origin='lower',vmin = 360,vmax=3500) #I don't really see why we need vmin or vmax in this scenario, but I think this works decently
plt.colorbar(label="Temperature (K)")
plt.xlabel("Pressure (bars)")
plt.ylabel("Volume (liters)")
plt.title("Van der Waals Gas Temperature")
plt.show() #The graph already looks pretty decent and cutting anything would lose some information
```

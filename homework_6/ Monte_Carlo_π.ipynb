```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#For this program, I referrenced the numpy documentation as well as geeksforgeeks for plt stuff
N = 10**3
x = np.random.rand(N)
y = np.random.rand(N)
count = 0
for i in range(len(x)):
    if (x[i]**2+y[i]**2) <= 1:
        count += 1 
pi = 4*(count/N)
print(pi)
```

```python
N = 10**4
x = np.random.rand(N)
y = np.random.rand(N)
within_x = []
within_y = []
outside_x=[]
outside_y = []
count = 0
for i in range(len(x)):
    if (x[i]**2+y[i]**2) <= 1:
        count += 1 
        within_x.append(x[i])
        within_y.append(y[i])
    else:
        outside_x.append(x[i])
        outside_y.append(y[i])
pi = 4*(count/N)
print(pi)
plt.figure()
plt.scatter(within_x,within_y,color="blue",label="within quarter circle")
plt.scatter(outside_x,outside_y,color="red",label="outside quarter circle")
plt.xlabel("X value")
plt.ylabel("Y value")
plt.title("Monte Carlo pi estimation")
plt.text(1.1,0,"pi value:"+str(pi),fontsize = 10) #I decided to put the value of pi off the graph so it shows up, if it was in the square, it would be covered
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#For this program I referenced the numpy documentation and geeksforgeeks for plt keywords
```

```python
x = np.linspace(1,40,40)
```

```python
y1 = np.random.randint(0,100,40)
y2 = np.random.randint(0,100,40)
y3 = np.random.randint(0,100,40)
```

```python
plt.figure()
fig, axs = plt.subplots(2,1)
axs[0].plot(x,y1,linewidth=10, color="orange",label="list 1")
axs[0].plot(x,y2, color="red",linestyle="dashed", label = "list 2")
axs[0].set_xlabel("Indexes")
axs[0].set_ylabel("Values")
axs[0].set_title("Plot of random numbers and its indexes")
axs[0].legend(loc="best")
axs[1].scatter(x,y3,marker="D",c="magenta")
axs[1].set_xlabel("Indexes")
axs[1].set_ylabel("Values")
axs[1].set_title("Plot of List 3 as a scatter plot")
plt.tight_layout()
plt.show() #This for some reason does not remove the <figure size 640x480 with 0 Axes, not sure why>
```

```python
from astropy.table import Table
import numpy as np
import pandas as pd
import scipy.optimize as opt
import matplotlib.pyplot as plt
```

```python
table = Table.read('global_CCl4_MM.dat',format = 'ascii')
data = table.to_pandas()
display(data)
```

```python
def time(year,month):
    dec_date = year + (month/12)
    return dec_date
date = time(data['CCl4ottoyr'],data['CCl4ottomon'])
data.insert(2,"Date",date)
data = data[['Date','CCl4ottoGLm','CCl4ottoGLsd']]
display(data)
```

```python
data = data.dropna()
std_error = data['CCl4ottoGLsd']/np.sqrt(len(data['CCl4ottoGLsd']))

plt.figure()
plt.errorbar(data['Date'],data['CCl4ottoGLm'],yerr = data['CCl4ottoGLsd'],ecolor = 'red',linewidth = 0.7) #Here I just used standard deviation as error instead of standard error
plt.xlabel("Date")
plt.ylabel("Global Mean of CCl4")
plt.title("The Global mean of Carbon Tetrachloride over the years")
plt.show()
```

```python
#fit
def f(x,m,b):
    return m*x + b

p0 = [1,1]
parameters, covariance = opt.curve_fit(f,data['Date'],data['CCl4ottoGLm'],p0)
```

```python
plt.figure()
plt.plot(data['Date'],f(data['Date'],parameters[0],parameters[1]),linestyle = 'dashed',label = 'fitted data')
plt.plot(data['Date'],data['CCl4ottoGLm'],label = 'actual data')
plt.xlabel("Date")
plt.ylabel("Global Mean of CCl4")
plt.title("The Global mean of Carbon Tetrachloride over the years with fitted curve")
plt.legend()
plt.show()
```

```python
#Statistical stuffs
residuals = data['CCl4ottoGLm'] - f(data['Date'],parameters[0],parameters[1])
chi = sum((residuals/data['CCl4ottoGLsd'])**2)
reduced = chi/(len(data['CCl4ottoGLm'])-1)

error_m = np.sqrt(covariance[0,0])
error_b = np.sqrt(covariance[1,1])
print('m={0:1.5f} +/- {1:1.5f}'.format(parameters[0],error_m))
print('b={0:1.5f} +/- {1:1.5f}'.format(parameters[1],error_b))
print() #empty line

print('f={0:1.5f}x + {1:1.5f}'.format(parameters[0],parameters[1]))
print('Reduced chi-squared = {0:1.5f}'.format(reduced))

plt.figure()
plt.scatter(data['Date'],residuals,s=2,c='blue')
plt.xlabel("Date")
plt.ylabel("Residuals")
plt.title("Residuals from fitting")
plt.show()
```

```python
print("The linear fit isn't bad, but it could be better. The residuals does show some kind of sinusoidal structure near the beginning, and the fitting was not perfect at times. However, just looking at the data, a linear fit works decently well.")
print() #empty line
print("Carbon Tetrachloride has be in decline because it depletes the ozone layer and is a greenhouse gas. Also, carbon tetrachloride is suspected to be a carcinogen so industries were forced to cut down on how much of it they produced.")
```

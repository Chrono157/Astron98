```python
import numpy as np
import matplotlib.pyplot as plt

def fourCurve():
    x = np.linspace(0,3,100)
    yerr = 0.05+2*x
    plt.figure()
    plt.plot(x,x**2,label='x^2')
    plt.plot(x,x**3,label='x^3')
    plt.plot(x,x**4,label='x^4')
    plt.plot(x,x**5,label = 'x^5')
        
    plt.title('Exponential Plots')
    plt.xlabel('X Values')
    plt.ylabel('exponents')
    plt.legend()
    plt.show()
    return

def fourCurveLoop():
    x = np.linspace(0,3,100)
    yerr = 0.05+2*x
    plt.figure()
    for i in range(2,6):
        #plt.plot(x,x**i,label='x^'+str(i))
        plt.errorbar(x,x**i,yerr=yerr,Label= 'x^'+str(i))
        
    plt.title('Exponential Plots')
    plt.xlabel('X Values')
    plt.ylabel('exponents')
    plt.legend()
    plt.show()
    return
fourCurveLoop()
```

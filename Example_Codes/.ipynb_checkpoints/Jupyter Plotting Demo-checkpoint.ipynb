```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(0,100)
y = np.sin(x)

plt.figure()
plt.plot(x,y,color='red',alpha=0.5)
plt.title('sine')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x2 = np.linspace(0,2*np.pi,100)
y2 = x2**2
```

```python
plt.figure(figsize=(7.3,10))
plt.plot(x2,y2,color='deepskyblue',marker='2',linestyle='dashdot',linewidth=20)
plt.title('figure 2')
plt.xlabel('x2')
plt.ylabel('y2')
plt.show()
```

```python
fig, axs = plt.subplots(2,2)
axs[0,0].plot(x,y)
axs[0,1].plot(x,y2)
axs[1,0].plot(x2,y)
axs[1,1].plot(x2,y2)
plt.show()
```

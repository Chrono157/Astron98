```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
x, y = np.meshgrid(np.linspace(-1,1, 100),np.linspace(-2,2,100))
z = x+y

plt.figure()
plt.imshow(z,cmap = "ocean")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.colorbar()
plt.show()
```

```python
x, y = np.meshgrid(np.linspace(-5,5, 100),np.linspace(-5,5,100))
z = (x**2+y**2)**0.5

plt.figure()
plt.imshow(z,cmap = "ocean")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.colorbar()
plt.show()
```
